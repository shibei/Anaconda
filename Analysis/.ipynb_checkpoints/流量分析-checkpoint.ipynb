```python
def get_data(date):#从sql中拉取某日的数据
    conn = psql.connect(host = 'localhost', user = 'root' , password = 'admin123' , database = 'device' )
    cursor = conn.cursor()
    sql = "SELECT * FROM `dev_info` WHERE `date` = "+ date
    df = pd.read_sql(sql,conn,)
    conn.close()
    return pd.DataFrame(df)

def groupd(a): #计算分组值
    group_num = 20
    groups = []
    #     print(a,min , max ,step)
#     print(bb)
    try:
        min = math.floor(bb['downSpeed'].min())
        max = math.ceil(bb['downSpeed'].max())
        step = math.ceil((max - min) / group_num)
        group_range = np.arange(min, max, step)
        for x in group_range:
            sub_group = [x , x+step]
            groups.append(sub_group)
        k = 0
        for i in range(group_num):
            if a >= groups[i][0] and a < groups[i][1]:
                return i
    except:
        pass

def tick(bb): #剔除离群值
    cc = bb.groupby('group').size().sort_values()
    y =0
    i = 0
    indexs = []
    for x in cc:
        y = y+x
        if y / cc.sum() > 0.1:
            indexs.append(cc.index[i])
        i = i + 1
    dd = indexs
    cc = cc.reindex(dd)
    ee = pd.DataFrame(cc).reset_index()
    ff = pd.merge(bb, ee, on='group').sort_values('clock').reset_index(drop=True)
    return ff,cc
```

```python
def clean_data(aa):#清洗数据
    high = []
    gwids = aa['gwid'].drop_duplicates().reset_index(drop=True)
#     print(gwids)
    for gwid in gwids:
        global bb 
        bb = aa[(aa['gwid']==gwid) & (aa['status']=='1')].reset_index(drop=True)
        grouped = bb['downSpeed'].apply(groupd)
        bb['group'] = grouped
        ff,p= tick(bb)
        p = p.sort_index()
        try:
            rate =p[p.index[-2:]].sum()/p.sum()
            if rate >= 0.20 and rate < 1 and len(p)>10:
                try : 
#                     print(p)
                    print(gwid,'lostData=',1-len(ff)/len(bb),bb['status'][0],rate) 
                except:
                    pass
#                 plt.hist(ff['downSpeed'])
#                 plt.show()
#                 plt.plot(ff['downSpeed'])
#                 plt.show()
                high.append(gwid)
        except:
            pass
    return high
```

```python
def getYesterday(x): 
    today=datetime.date.today() 
    oneday=datetime.timedelta(days=x) 
    yesterday=today-oneday  
    return yesterday
```

```python
h = []
for i in range(7,0,-1):
    date = "'"+str(getYesterday(i))+"'"
    print(date)
    aa = get_data(date)
    h.append(clean_data(aa))
```
